```python
def ANOVA(df,cat,con):
    from statsmodels.formula.api import ols
    rel=con+ "~"+cat
    model=ols(rel,df).fit()
    from statsmodels.stats.anova import anova_lm
    anova_results=anova_lm(model)
    return round(anova_results.iloc[0,4],4)
```

```python
imp_con_cols=[]
for i in cat:
    print("---------------")
    print("--------MonthlyRate vs",i,"---------")
    x=ANOVA(A,i,"MonthlyRate")
    print(x)
    if x<0.05:
        imp_cat_cols.append(i)
```

# Create regression models:
    1.SLR
    2.MLR
    3.DTR
    4.RFR|ABR

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import mean_absolute_error,explained_variance_score
```

```python
def model(mobj):
    model=mobj.fit(xtrain,ytrain)
    pred_tr=model.predict(xtrain)
    pred_ts=model.predict(xtest)
    tr_err=mean_absolute_error(ytrain,pred_tr)
    ts_err=mean_absolute_error(ytest,pred_ts)
    exp_var=explained_variance_score(ytest,pred_ts)
    return round(tr_err,3),round(ts_err,3),exp_var,mobj
```

```python
lm=LinearRegression()
model(lm)
```
